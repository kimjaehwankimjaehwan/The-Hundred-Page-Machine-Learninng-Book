```python
def update_w_and_b(spendings, sales, w, b, alpha):
    dl_dw = 0.0
    dl_db = 0.0
    N = len(spendings)

    for i in range(N):
        dl_dw += -2 * spendings[i] * (sales[i] - (w * spendings[i] + b))
        dl_db += -2 * (sales[i] - (w * spendings[i] + b))

    w = w - (dl_dw / N) * alpha
    b = b - (dl_db / N) * alpha

    return w, b
```

```python
def train(spendings, sales, w, b, alpha, epochs):
    for e in range(epochs):
        w, b = update_w_and_b(spendings, sales, w, b, alpha)

        # 전환 과정 출력
        if e % 400 == 0:
            print("epoch:", e, "loss:", avg_loss(spendings, sales, w, b))

    return w, b
```

```python
def avg_loss(spendings, sales, w, b):
    N = len(spendings)
    total_error = 0.0
    for i in range(N):
        total_error += (sales[i] - (w * spendings[i] + b)) ** 2
    return total_error / float(N)
```

```python
def predict(x, w, b):
    return w * x + b
```

```python
x = [10, 20, 30, 40, 50]
y = [20, 40, 60, 80, 100]

w, b = train(x, y, 0.0, 0.0, 0.0001, 15000)
x_new = 23.0
y_new = predict(x_new, w, b)
print(y_new)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 손실 함수의 평균 제곱 오차(MSE)를 계산하는 함수
def avg_loss(spendings, sales, w, b):
    N = len(spendings)
    total_error = 0.0
    for i in range(N):
        total_error += (sales[i] - (w * spendings[i] + b)) ** 2
    return total_error / float(N)

# 경사 감소법을 이용해 파라미터 w와 b를 업데이트하는 함수
def update_w_and_b(spendings, sales, w, b, alpha):
    dl_dw = 0.0
    dl_db = 0.0
    N = len(spendings)

    for i in range(N):
        dl_dw += -2 * spendings[i] * (sales[i] - (w * spendings[i] + b))
        dl_db += -2 * (sales[i] - (w * spendings[i] + b))

    w = w - (dl_dw / N) * alpha
    b = b - (dl_db / N) * alpha

    return w, b

# 전체 에포크 동안 파라미터 w와 b를 학습시키는 함수
def train(spendings, sales, w, b, alpha, epochs):
    w_list = []
    b_list = []
    for e in range(epochs):
        w, b = update_w_and_b(spendings, sales, w, b, alpha)
        if e % 400 == 0:
            print(f"epoch: {e}, loss: {avg_loss(spendings, sales, w, b)}")
            w_list.append(w)
            b_list.append(b)
    return w, b, w_list, b_list

# 새로운 데이터에 대해 예측하는 함수
def predict(x, w, b):
    return w * x + b

# 데이터를 플롯하는 함수
def plot_regression(spendings, sales, w_list, b_list):
    # epochs = [0, 400, 800, 1200, 1600]  # This line is causing the issue
    epochs = [i * 400 for i in range(len(w_list))] # Generate epochs list based on the length of w_list
    fig, axs = plt.subplots(2, 3, figsize=(15, 10))

    # Access only the existing elements in w_list and b_list
    for i, ax in enumerate(axs.flat[:len(w_list)]):
        w = w_list[i]
        b = b_list[i]
        ax.scatter(spendings, sales, color="black")
        line_x = [min(spendings), max(spendings)]
        line_y = [predict(min(spendings), w, b), predict(max(spendings), w, b)]
        ax.plot(line_x, line_y, color="blue")
        ax.set_title(f"에포크 {epochs[i]}")
        ax.set_xlabel("광고비 지출")
        ax.set_ylabel("판매량")

    plt.tight_layout()
    plt.show()

# 실제 실행 예시

# 광고비 지출 데이터 (100개의 점 생성, 0~100 사이의 값)
np.random.seed(42)  # 재현성을 위해 시드 고정
spendings = np.linspace(0, 100, 100)

# 실제 판매량 데이터 (노이즈가 추가된 값)
true_slope = 0.65
true_intercept = 5
noise = np.random.normal(0, 5, size=spendings.shape)  # 평균 0, 표준편차 5인 노이즈 추가
sales = true_slope * spendings + true_intercept + noise

# 초기 파라미터 설정
w, b = 0.7, 0.1
alpha = 0.00001
epochs = 1600

# 학습 수행
w, b, w_list, b_list = train(spendings, sales, w, b, alpha, epochs)

# 학습 과정 시각화
plot_regression(spendings, sales, w_list, b_list)
```
