```python
def update_w_and_b(spendings, sales, w, b, alpha):
    dl_dw = 0.0
    dl_db = 0.0
    N = len(spendings)

    for i in range(N):
        dl_dw += -2 * spendings[i] * (sales[i] - (w * spendings[i] + b))
        dl_db += -2 * (sales[i] - (w * spendings[i] + b))

    w = w - (dl_dw / N) * alpha
    b = b - (dl_db / N) * alpha

    return w, b
```

```python
def train(spendings, sales, w, b, alpha, epochs):
    for e in range(epochs):
        w, b = update_w_and_b(spendings, sales, w, b, alpha)

        # 전환 과정 출력
        if e % 400 == 0:
            print("epoch:", e, "loss:", avg_loss(spendings, sales, w, b))

    return w, b
```

```python
def avg_loss(spendings, sales, w, b):
    N = len(spendings)
    total_error = 0.0
    for i in range(N):
        total_error += (sales[i] - (w * spendings[i] + b)) ** 2
    return total_error / float(N)
```

```python
def predict(x, w, b):
    return w * x + b
```

```python
x = [10, 20, 30, 40, 50]
y = [20, 40, 60, 80, 100]

w, b = train(x, y, 0.0, 0.0, 0.0001, 15000)
x_new = 23.0
y_new = predict(x_new, w, b)
print(y_new)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 손실 함수의 평균 제곱 오차(MSE)를 계산하는 함수
def avg_loss(spendings, sales, w, b):
    N = len(spendings)
    total_error = 0.0
    for i in range(N):
        total_error += (sales[i] - (w * spendings[i] + b)) ** 2
    return total_error / float(N)

# 경사 감소법을 이용해 파라미터 w와 b를 업데이트하는 함수
def update_w_and_b(spendings, sales, w, b, alpha):
    dl_dw = 0.0
    dl_db = 0.0
    N = len(spendings)

    for i in range(N):
        dl_dw += -2 * spendings[i] * (sales[i] - (w * spendings[i] + b))
        dl_db += -2 * (sales[i] - (w * spendings[i] + b))

    w = w - (dl_dw / N) * alpha
    b = b - (dl_db / N) * alpha

    return w, b

# 전체 에포크 동안 파라미터 w와 b를 학습시키는 함수
def train(spendings, sales, w, b, alpha, epochs):
    w_list = []
    b_list = []
    for e in range(epochs + 1):  # 에포크는 0부터 epochs까지 학습
        w, b = update_w_and_b(spendings, sales, w, b, alpha)
        if e % 400 == 0 or e == 1600 or e == 3000:  # 1600, 3000에서도 저장
            print(f"epoch: {e}, loss: {avg_loss(spendings, sales, w, b)}")
            w_list.append(w)
            b_list.append(b)
    return w, b, w_list, b_list

# 새로운 데이터에 대해 예측하는 함수
def predict(x, w, b):
    return w * x + b

# 데이터를 플롯하는 함수
def plot_regression(spendings, sales, w_list, b_list, epochs):
    fig, axs = plt.subplots(2, 3, figsize=(15, 10))

    # Access only the existing elements in w_list and b_list
    for i, ax in enumerate(axs.flat[:len(w_list)]):
        w = w_list[i]
        b = b_list[i]
        line_x = [min(spendings), max(spendings)]
        line_y = [predict(min(spendings), w, b), predict(max(spendings), w, b)]
        ax.scatter(spendings, sales, color="black")
        ax.plot(line_x, line_y, color="blue")
        ax.set_title(f"에포크 {epochs[i]}")
        ax.set_xlabel("광고비 지출")
        ax.set_ylabel("판매량")

    plt.tight_layout()
    plt.show()

# 실제 실행 예시

# 광고비 지출 데이터 (100개의 점 생성, 0~100 사이의 값)
np.random.seed(42)  # 재현성을 위해 시드 고정
spendings = np.linspace(0, 100, 100)

# 실제 판매량 데이터 (노이즈가 추가된 값)
true_slope = 0.65
true_intercept = 5
noise = np.random.normal(0, 10, size=spendings.shape)  # 평균 0, 표준편차 5인 노이즈 추가
sales = true_slope * spendings + true_intercept + noise

# 초기 파라미터 설정
w, b = 1.0, 0.1
alpha = 0.00001
epochs = 3000

# 학습 수행
w, b, w_list, b_list = train(spendings, sales, w, b, alpha, epochs)

# 에포크 정보 추가
epochs_list = [i * 400 for i in range(len(w_list) - 2)] + [1600, 3000]

# 학습 과정 시각화
plot_regression(spendings, sales, w_list, b_list, epochs_list)
```

```python
# 필요한 라이브러리 임포트
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# 모델 학습 함수 정의
def train(x, y):
    # 선형 회귀 모델 생성 및 학습
    model = LinearRegression().fit(x, y)
    return model

# 예측 함수 정의
def predict(model, x_new):
    # 학습된 모델을 사용하여 예측
    return model.predict([[x_new]])

# 시각화 함수 정의
def plot_regression_line(x, y, model):
    # 원본 데이터 산점도 그리기
    plt.scatter(x, y, color='black', label="Original Data")

    # 학습된 회귀선을 위한 x축 값 생성
    line_x = np.linspace(min(x), max(x), 100)
    line_y = model.predict(line_x.reshape(-1, 1))

    # 회귀선 그리기
    plt.plot(line_x, line_y, color='blue', label="Regression Line")

    # 그래프 제목 및 라벨 추가
    plt.title("Linear Regression with Noise")
    plt.xlabel("Advertising Spendings")
    plt.ylabel("Sales")
    plt.legend()

    # 그래프 출력
    plt.show()

# 데이터 생성 (광고비 지출과 노이즈가 포함된 판매량 데이터)
np.random.seed(42)  # 재현성을 위해 시드 고정
x = np.linspace(0, 100, 100).reshape(-1, 1)  # 광고비 지출 데이터 (100개의 점)
true_slope = 0.65
true_intercept = 5
noise = np.random.normal(0, 9, size=x.shape)  # 평균 0, 표준편차 5인 노이즈 추가
y = true_slope * x + true_intercept + noise  # 실제 판매량 데이터에 노이즈 추가

# 모델 학습
model = train(x, y)

# 새로운 데이터 예측 (광고비 지출 23에 대한 예측)
x_new = 23.0
y_new = predict(model, x_new)

# 예측 결과 출력
print(f"광고비 지출 {x_new}에 대한 예상 판매량: {y_new[0]}")

# 회귀선 및 데이터 시각화
plot_regression_line(x, y, model)
```
